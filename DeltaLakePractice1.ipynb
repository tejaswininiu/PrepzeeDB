```python
from delta.tables import *
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# Create Sample Data

```python
data = [
    (1,"John Doe", "Engineering", 75000,"2023-01-15"),
    (2,"Jane Smith", "Marketing", 60000,"2023-02-20"),
    (3,"Mike Johnson", "Engineering", 80000,"2023-03-01"),
    (4,"Emily Davis", "Sales", 55000,"2023-04-05"),
    (5,"David Lee", "Engineering", 70000,"2023-05-10"),
    (6,"Olivia Garcia", "Marketing", 65000,"2023-06-15")
]
```

## Schema Definition

```python
schema =StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("department", StringType(), True),
    StructField("salary", IntegerType(), True),
    StructField("hire_date", StringType(), True)
])
```

## create dataframe

```python
df= spark.createDataFrame(data, schema)
df.show()
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("employees_delta")
print("Delta table created successfully")
```

```python
df_read = spark.read.table("employees_delta")
df_read.show()
```

Read Delta table(SQL example)

```python
%sql
SELECT * from employees_delta
```

Insert records into Delta Table


```python
new_data = [
    (7,"Sophia Rodriguez", "Engineering", 72000,"2023-07-20"),
    (8,"Daniel Kim", "Marketing", 68000,"2023-07-20")
]
new_df = spark.createDataFrame(new_data, schema)
new_df.show()
new_df.write.format("delta").mode("append").saveAsTable("employees_delta")
```

```python
%sql
Select * from employees_delta
```
